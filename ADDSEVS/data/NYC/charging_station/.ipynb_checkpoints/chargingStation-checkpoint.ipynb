```python
cs_shp = gpd.read_file('NYC_zone_charger/NYC_zone_wgs84_l1_l2_dc_charger.shp')
```

```python
cs_shp = cs_shp[cs_shp['num']>0]
```

```python
cs_shp['geometry'] = cs_shp['geometry'].centroid
```

```python
cs_shp.to_file('charging_station_real.shp')
```
